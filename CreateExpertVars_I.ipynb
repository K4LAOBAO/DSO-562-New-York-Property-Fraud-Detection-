```python
cleanDF = pd.read_csv('Combined After Fill Missing.csv')
cleanDF.columns = [x.upper() for x in cleanDF.columns]
tmp = cleanDF.describe()

cleanDF.info()
```

```python
cols = ["RECORD",
        "FULLVAL",
        "AVLAND",
        "AVTOT",
        "LTFRONT",
        "LTDEPTH",
        "BLDFRONT",
        "BLDDEPTH",
        "STORIES",
        "ZIP",
        "TAXCLASS",
        "B",            
       ]
```

```python
cl_DF = cleanDF[cols]
```

```python
cl_DF.info()
```

```python
cl_DF["S1"] = cl_DF.LTFRONT*cl_DF.LTDEPTH
cl_DF["S2"] = cl_DF.BLDFRONT*cl_DF.BLDDEPTH
cl_DF["S3"] = cl_DF.S2*cl_DF.STORIES

#cl_DF["R1"] = cl_DF./(cl_DF.)
```

```python
cl_DF["R1"] = cl_DF.FULLVAL/cl_DF.S1
cl_DF["R2"] = cl_DF.FULLVAL/cl_DF.S2
cl_DF["R3"] = cl_DF.FULLVAL/cl_DF.S3

cl_DF["R4"] = cl_DF.AVLAND/cl_DF.S1
cl_DF["R5"] = cl_DF.AVLAND/cl_DF.S2
cl_DF["R6"] = cl_DF.AVLAND/cl_DF.S3

cl_DF["R7"] = cl_DF.AVTOT/cl_DF.S1
cl_DF["R8"] = cl_DF.AVTOT/cl_DF.S2
cl_DF["R9"] = cl_DF.AVTOT/cl_DF.S3
```

```python
cl_DF["ZIP3"] = round(cl_DF["ZIP"]/100)
#cl_DF['ZIP3'] = cl_DF['ZIP'].apply(lambda x: x[:3])
```

```python
groups = ['ZIP', 
        'ZIP3' ,
        'TAXCLASS', 
        'B']

VarsR = ['R1','R2','R3','R4','R5','R6','R7','R8','R9']
```

```python
# Create 36 new Expert Columns
#==============================
i=1
for g in groups:
    for v in VarsR:
        #for g in groups
        #for v in VarsR
        newCol = g+"_avgd_"+v
        print('This is ',i,' -> ',newCol)
        #cl_DF[newCol]=cl_DF[[v]].div(cl_DF.groupby(g).transform('mean'))
        #cl_DF[newCol]=cl_DF[v].div(cl_DF.groupby(cl_DF[g]).transform('mean'))
        cl_DF[newCol]=cl_DF[[v]]/(cl_DF[[v,g]].groupby(g).transform('mean'))
        i=i+1
```

```python
#Create the last 9 Expert Variables
#===============================
i=37
for v2 in VarsR:
    newCol = "All_avgd_"+v2
    print('This is ',i,' -> ',newCol)
    cl_DF[newCol]=cl_DF[v2]/cl_DF[v2].mean()
    i=i+1
```
