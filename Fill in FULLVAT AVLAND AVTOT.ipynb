```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
%matplotlib inline
```

```python
data=pd.read_csv('NY property data.csv')
```

```python
zipstory=pd.read_csv("storyZip2.csv")
```

```python
zipstory["ZIP5"].value_counts()
```

```python
merged=pd.concat([data,zipstory],axis=1)
```

```python
merged.drop('Unnamed: 0', axis=1, inplace=True)
```

```python
merged
```

## FULLVAL

```python
FULLVAL=merged[["FULLVAL","ZIP5"]].copy()
FULLVAL["FULLVALnew"]=FULLVAL["FULLVAL"]
FULLVAL['FULLVALnew'].replace(0, np.nan, inplace= True)
```

```python
FULLVAL.drop(["FULLVAL"],axis=1,inplace=True)
```

```python
FULLVAL["FULLVALnew"] = FULLVAL.groupby("ZIP5").\
   transform(lambda x: x.fillna(x.median()))
```

```python
FULLVAL1=FULLVAL["FULLVALnew"]
```
